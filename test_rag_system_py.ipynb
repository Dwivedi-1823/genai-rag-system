# Chunking

```python
!pip install langchain
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500, chunk_overlap=50, separators=["\n\n", "\n", " ", ""]
)
```

```python
chunks = splitter.create_documents([content[0].page_content])
```

### Initialize the Generative Model (LLM)

```python
import os

# Ensure GOOGLE_API_KEY is set in the environment before this cell runs.
# Check if the API key is present
if "GOOGLE_API_KEY" not in os.environ:
    print("ERROR: GOOGLE_API_KEY environment variable not set. Please ensure cell e937dd97 is executed successfully.")
    # You might want to raise an error here or handle it as appropriate
    raise ValueError("GOOGLE_API_KEY not found in environment variables.")
else:
    print("GOOGLE_API_KEY found in environment variables.")

llm = ChatGoogleGenerativeAI(model="gemini-pro")
```

### Define a Prompt Template

```python
prompt = ChatPromptTemplate.from_template(
    """Answer the following question based only on the provided context:

    <context>
    {context}
    </context>

    Question: {input}"""
)
```

### Create a Document Combination Chain

```python
document_chain = create_stuff_documents_chain(llm, prompt)
```

### Create a Retrieval Chain

```python
retrieval_chain = create_retrieval_chain(retriver, document_chain)
```

### Invoke the Retrieval Chain with a Query

```python
response = retrieval_chain.invoke({"input": query})
print(response["answer"])
```
